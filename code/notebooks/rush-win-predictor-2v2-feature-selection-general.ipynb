```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from make_plot_dir import plots_dir
from notebook_function_space import get_multicollinear_sets, remove_multicollinear_features
```

```python
pro_frames_merge = pd.read_csv('../../data/pro_frames_merge_2v2.csv', index_col = 0)
not_ol = ['HB', 'WR', 'FB', 'TE']
pro_frames_merge = pro_frames_merge[(~pro_frames_merge['position_in'].isin(not_ol)) &
                                    (~pro_frames_merge['position_out'].isin(not_ol))]
pro_frames_merge['stunt_type'] = pro_frames_merge['stunt_type'].astype('category')
```

```python
features = ['width_in','width_out', 
            'depth_in', 'depth_out',
            's_in','s_out',
            'a_in', 'a_out',
            'squareness_in', 'squareness_out',
            'open_outside_in', 'open_outside_out', 
            'moving_outside_in', 'moving_outside_out',
            'rotation_outside_in', 'rotation_outside_out',
            'qb_squareness_in', 'qb_squareness_out',
            'qb_dist_in', 'qb_dist_out',
            'x_diff', 'y_diff', 'dist', 'min_qb_dist', 'rel_rotation', 'penetrator_depth', 
            'harm_squareness','mean_squareness', 'rms_squareness', 'max_squareness',
            'harm_qb_squareness','mean_qb_squareness', 'rms_qb_squareness', 'max_qb_squareness',
            'stunt_type', 'exchange']
```

```python
delta_cols = []
for col in features[:-2]:
    lag_col = pro_frames_merge.groupby(['gameId', 'playId', 'stuntId'])[col].shift(1)
    delta_col = col + '_delta'
    delta_cols.append(delta_col)
    pro_frames_merge[delta_col] = pro_frames_merge[col] - lag_col
features = features + delta_cols
features
```

```python
frames_query = "frame_from_snap > 0 & frame_from_overlap < 11 & stunt_type != 'EE'"
pro_frames_merge_trunc = pro_frames_merge.query(frames_query).copy()
num_weeks = pro_frames_merge_trunc['week'].max()
y = pro_frames_merge_trunc['rush_win']
y.mean()
```

```python
pro_frames_merge_trunc.groupby('week')['rush_win'].mean()
```

```python
df = pro_frames_merge_trunc[features + ['rush_win']]
corr_matrix = df.corr(numeric_only = True)
win_corr = corr_matrix['rush_win'][:-1]
win_corr.sort_values(ascending = False)
```

```python
win_corr.abs().sort_values(ascending = False)[:50]
```

```python
corr_sets = get_multicollinear_sets(pro_frames_merge_trunc,features, threshold = 0.7)
corr_sets
```

```python
chosen_features = [
                   'penetrator_depth',
                   'mean_qb_squareness',
                   'mean_qb_squareness_delta',
                   'moving_outside_out', 
                   's_in', 's_out',
                     'min_qb_dist',
                     'min_qb_dist_delta',
                     'x_diff',
                     'squareness_out',
                     'squareness_out_delta',
                     'squareness_in_delta'
                  ]
remaining_sets, features2 = remove_multicollinear_features(features, corr_sets, chosen_features)
remove_features = ['qb_squareness_in_delta', 
                   'harm_squareness_delta',
                   'max_squareness_delta']
features2 = [feat for feat in features2 if feat not in remove_features]
features2
```

```python
corr_sets2 = get_multicollinear_sets(pro_frames_merge_trunc,features2, threshold = 0.5)
corr_sets2
```
